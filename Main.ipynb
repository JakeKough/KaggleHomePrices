Import Packages to be used

```python
import pandas as pd
import sklearn
import numpy as np

```

read data

```python
print("---------------------Read Data-----------------------")
train_data_full = pd.read_csv("train.csv")
test_data_full = pd.read_csv("test.csv")
print("---Training Data---")
print(train_data_full.head())
print("\n---Testing Data---")
print(test_data_full.head())
print(test_data_full.info())
test_data_ids = test_data_full['Id']
print("\n---Test Data IDs---")
print(test_data_ids.head())
```

Drop rows where the value to predict (Sales Price) is null

```python
train_data_full.dropna(axis=0, subset= ['SalePrice'], inplace = True)
```

Seperate Features from Target prediction

```python
y = train_data_full['SalePrice']
X = train_data_full.drop(['SalePrice'],axis = 1)
```

Split Test dataset test and validation sets for training

```python
from sklearn.model_selection import train_test_split

train_x,val_x,train_y,val_y = train_test_split(X,y,train_size = .8, test_size = .2, random_state = 1)
train_x.head()
```

investigate the dataset

```python
def investigate(df):
    print("---columns---")
    print(df.columns)
    print("---describe---")
    df.describe()
    print("---shape---")
    print(df.shape)
    print("---data types---")
    print(df.dtypes)
    print("---info---")
    df.info()
```

```python
investigate(train_x)
```

Decide Features to keep initially and keep features in training dataset

```python
features = ['MSSubClass','MSZoning','LotFrontage','LotArea','Street','Alley', 'Utilities', 'BldgType',
       'HouseStyle', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinSF1','BsmtFinType2', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'Heating',
       'HeatingQC', 'CentralAir', 'Electrical', '1stFlrSF', '2ndFlrSF','LowQualFinSF', 'GrLivArea', 'FullBath',
       'HalfBath', 'KitchenQual','TotRmsAbvGrd', 'Fireplaces', 'GarageType','GarageYrBlt', 'GarageFinish', 'GarageCars', 'GarageArea', 'GarageQual',
       'GarageCond', 'PavedDrive', 'WoodDeckSF', 'OpenPorchSF','EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'PoolQC','Fence', 'SaleType', 'SaleCondition','BedroomAbvGr']

def passfeatures(df,feat):
    return df[feat]

train_x = passfeatures(train_x,features)
val_x = passfeatures(val_x,features)
test_data_full = passfeatures(test_data_full,features)
```

```python
#Engineer Fields
import math
from datetime import datetime

def engineer(df):
    df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
    df['hasGarage'] = df['GarageYrBlt'].apply(lambda x: 0.0 if pd.isnull(x) else 1.0)
    df['hasBasement'] = df['TotalBsmtSF'].apply(lambda x: 0.0 if x == 0 else 1.0)
    df['TotalBath'] = df['FullBath'] + df['HalfBath']
    df['hasFireplace'] = df['Fireplaces'].fillna(0).apply(lambda x: 0.0 if pd.isnull(x) or x < 1 else 1.0)
    df['Remodeled']= df['YearRemodAdd'].apply(lambda x: 0.0 if pd.isnull(x) else 1.0)
    df['Age'] = df['YearBuilt'].apply(lambda x: datetime.now().year - x)

engineer(train_x)
engineer(val_x)
engineer(test_data_full)
train_x.info()
```

```python
#normalize the int and float fields using min max
from sklearn.preprocessing import MinMaxScaler

print(train_x.head())
number_fields = []
for col in train_x:
    if str(train_x.dtypes[col]) in ["int64","float64"]:
        number_fields.append(col)

print(number_fields)

scaler = MinMaxScaler()
train_x[number_fields] = scaler.fit_transform(train_x[number_fields])
val_x[number_fields] = scaler.transform(val_x[number_fields])
test_data_full[number_fields] = scaler.transform(test_data_full[number_fields])
print(train_x.head())
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

print(train_x.head())
number_fields_not_null = []
number_fields_with_null = []
for col in train_x:
    if train_x.dtypes[col] in ["int64","float64"] and not train_x[col].isnull().values.any():
        number_fields_not_null.append(col)
    if train_x.dtypes[col] in ["int64","float64"] and train_x[col].isnull().values.any():
        number_fields_with_null.append(col)

print(number_fields_not_null)
print(number_fields_with_null)

model = GradientBoostingRegressor()
model.fit(train_x[number_fields_not_null], train_y)
predictions = model.predict(val_x[number_fields_not_null])



score = r2_score(val_y,predictions)
mae = mean_absolute_error(val_y, predictions)
print(score)
print("MAE: ", mae)
```

```python
# Run the model test set

for col in test_data_full[number_fields_not_null]:
    test_data_full[col] = test_data_full[col].fillna(0)
print(test_data_full[number_fields_not_null].head())
test_data_full[number_fields_not_null].info()

preds_test = model.predict(test_data_full[number_fields_not_null])


output = pd.DataFrame({'Id': test_data_ids,
                       'SalePrice': preds_test})

print(output.head())
output.to_csv('submission.csv', index=False)
```
